```python
orders = pd.read_csv('2021.csv', header = None)
orders.head()
```

```python
#Check for duplicates
duplicates = orders[orders.duplicated()]

if not duplicates.empty:
    print("Duplicate rows: ", duplicates)
else:
    print("No duplicates")
```

```python
#Add Column Names
col_names = ['order_id', 'line', 'order_date', 'customer_name', 'email_address', 'product', 'quantity', 'unit_price', 'tax_amount']
```

```python
#Read col_names into the csv file
orders = pd.read_csv('2021.csv',  header = None, names = col_names)
orders.head()
```

```python
orders.columns.value_counts().sum()
```

```python
orders.shape
```

```python
orders.columns.value_counts().sum()
```

```python
#Splitting the order date into year, month and day
orders[['year', 'month', 'day']] = orders['order_date'].str.split('-', expand = True)

orders.head()
```

```python
#Get order date index
date_index = orders.columns.get_loc('order_date')

#Insert Year and Month column after the Order date
orders.insert(date_index + 1, 'year', orders.pop("year"))
orders.insert(date_index + 2, 'month', orders.pop("month"))

#Cross check
orders.head()
```

```python
#Drop the day column
orders.drop('day', axis = 1, inplace = True)

orders.head()
```

```python
#Converting object to datetime format
import datetime as dt

orders['order_date'] = [dt.datetime.strptime(x, "%Y-%m-%d")
                            for x in orders['order_date']]
```

```python
orders['order_date'].dtype
```

```python
orders[['first_name', 'last_name']] = orders['customer_name'].str.split(' ', n = 1, expand = True)

orders.head()
```

```python
#Separte the customer name into first and last name
name_index = orders.columns.get_loc('customer_name')

#Insert first name and last name column before the customer name column
orders.insert(name_index + 0, 'first_name', orders.pop("first_name"))
orders.insert(name_index + 1, 'last_name', orders.pop("last_name"))

orders.head()
```

```python
orders.drop('customer_name', axis = 1, inplace = True)

orders.head()
```

```python
#Save to file
orders.to_csv("orders.csv", index = False)
```
